## 1) Functional Method

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import randint
import pandas as pd
```

```python
x = np.linspace(0,10,20)
x
```

```python
y = randint(0,50,20)
y
```

```python
y = np.sort(y)
y
```

```python
plt.plot(x,y,color='c',linestyle='--',marker='^')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('X vs Y axis')
plt.show()
```

## Multiple plots on some canvas

```python
plt.subplot(1,2,1)
plt.plot(x,y,color='r')
plt.subplot(1,2,2)
plt.plot(x,y,color='m')
```

## 2 sets of figures to one canvas

```python
fig = plt.figure()
ax1=fig.add_axes([0.1,0.1,0.8,0.8])
ax2=fig.add_axes([0.2,0.5,0.4,0.3])    # parameters left,bottom,width,height

ax1.plot(x,y,color='r')
ax1.set_xlabel('X axis')
ax1.set_ylabel('Y axis')
ax1.set_title('Plot 1')

ax2.plot(x,y,color='m')
ax2.set_xlabel('X axis')
ax2.set_ylabel('Y axis')
ax2.set_title('Plot 2')
```

```python
fig,axes=plt.subplots()
axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=3)
plt.tight_layout() # for creating space among each other
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
for current_ax in axes:
    current_ax.plot(x,y)
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[1].plot(x,y)
axes[0].set_title('Plot 1')
axes[1].set_title('Plot 2')
```

```python
plt.scatter(days,y_views,label='Youtube',marker='o')
plt.scatter(days,f_views,label='Facebook',marker='o')
plt.scatter(days,t_views,label='Twitter',marker='o')
plt.xlabel('Days')
plt.ylabel('Views')
plt.title('Social Media views')
plt.legend()
```

```python
plt.scatter(days,y_views,label='Youtube',marker='o')
plt.scatter(days,f_views,label='Facebook',marker='o')
plt.scatter(days,t_views,label='Twitter',marker='o')
plt.xlabel('Days')
plt.ylabel('Views')
plt.grid(color='r',linestyle='--')
plt.title('Social Media views')
plt.legend()
```

 ## 2) Bar plot

```python
plt.bar(days,y_views,label='Youtube Videos')
plt.bar(days,f_views,label='Facebook Videos')
plt.xlabel('Days')
plt.ylabel('Views')
plt.title('Social Media views')
plt.legend()
```
